```python
import numpy as np

J = 1.0
h = 1.0

# Define Pauli matrices
X = np.array([[0, 1], [1, 0]])  # Pauli-X matrix
Y = np.array([[0, -1j], [1j, 0]])  # Pauli-Y matrix
Z = np.array([[1, 0], [0, -1]])  # Pauli-Z matrix
I = np.eye(2)  # Identity matrix

def XX_term(N, index):
    if index < N - 1:
        return J * np.kron(
            np.eye(2**index), np.kron(np.kron(X, X), np.eye(2 ** (N - index - 2)))
        )
    if index == N - 1:
        return J * np.kron(np.kron(X, np.eye(2**(N - 2))), X)

def Z_term(N, index):
    if index <= N - 1:
        return h * np.kron(np.eye(2**index), np.kron(Z, np.eye(2 ** (N - index - 1))))
```

```python
sum = 0
def commutator(A, B):
    # spectral norm
    return np.linalg.norm(A @ B - B @ A, ord=2)

def nested_commutator(N):
    sum = 0
    for i in range(N):
        for j in range(N):
            sum += commutator(XX_term(N, i), XX_term(N, j))
            sum += commutator(XX_term(N, i), Z_term(N, j))
            sum += commutator(Z_term(N, i), Z_term(N, j))
            sum += commutator(Z_term(N, i), XX_term(N, j))
    return sum
```

```python
print(nested_commutator(9))
```

```python
res = []
N_list = np.arange(4, 11)
for N in N_list:
    res.append(nested_commutator(N))
    
# linear regression
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

logN = np.log(N_list)
logres = np.log(res)


k, b = np.polyfit(logN, logres, 1)
x = np.linspace(min(logN), max(logN), 100)

ax.plot(logN, logres, "o", color=f"C0")
ax.plot(x, k * x + b, color=f"C0", label=f"Slope: {k:.2f}")

ax.set_xlabel("log(N)")
ax.set_ylabel("log(Commutator Norm)")
ax.set_title("Scaling of Commutator Norm with System Size N")
ax.legend()
plt.show()
```
